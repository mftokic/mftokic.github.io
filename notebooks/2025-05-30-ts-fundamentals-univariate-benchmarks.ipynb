### Python Code

```python
# import libraries
import pandas as pd
from statsforecast import StatsForecast
from statsforecast.models import WindowAverage, Naive, SeasonalNaive
import matplotlib.pyplot as plt
```

```python
# read data
data_raw = pd.read_csv("../posts/2024-10-02-ts-fundamentals-whats-a-time-series/example_ts_data.csv")

data_raw = (
    # select columns
    data_raw[["Country", "Product", "Date", "Revenue"]]
    # change data types
    .assign(
        Date = pd.to_datetime(data_raw["Date"]), 
        Revenue = pd.to_numeric(data_raw["Revenue"])
    )
)

# print the first few rows
print(data_raw.head())
```

```python
# filter on specific series
us_ic_raw = data_raw[(data_raw["Country"] == "United States") & (data_raw["Product"] == "Ice Cream")]

# create unique id
us_ic_raw["unique_id"] = us_ic_raw["Country"] + "_" + us_ic_raw["Product"]

# convert date to datetime
us_ic_raw["Date"] = pd.to_datetime(us_ic_raw["Date"])

# plot the data
plt.figure(figsize=(10, 6))
plt.plot(us_ic_raw.index, us_ic_raw["Revenue"], label="Ice Cream Revenue")
plt.xlabel("Date")
plt.ylabel("Revenue")
plt.title("Ice Cream Revenue in United States")
plt.legend()
```

```python
# get final data for forecasting
us_ic_clean = us_ic_raw[["unique_id", "Date", "Revenue"]].copy()

# set up models to train
models = [
    WindowAverage(window_size = 12),          # meanf
    Naive(),                                # naive
    SeasonalNaive(season_length=12),        # snaive
]

sf = StatsForecast(
    models=models,
    freq='ME'
)

# fit the model and forecast for 12 months ahead
Y_hat_df = sf.forecast(df = us_ic_clean, 
                       time_col = "Date", 
                       target_col = "Revenue", 
                       id_col = "unique_id", 
                       h=12, 
                       fitted=False)

print(Y_hat_df.head())

# convert date to be first of the month
Y_hat_df["Date"] = Y_hat_df["Date"].dt.to_period("M").dt.to_timestamp()
```

```python
# concat both df together
combined_df = pd.concat([us_ic_clean, Y_hat_df], axis=0)

# make date the index
combined_df.set_index("Date", inplace=True)

print(combined_df.head())
```

```python
# plot the combined data
plt.figure(figsize=(10, 6))

# plot the original revenue data as line and forecast as dotted line
plt.plot(combined_df.index, combined_df["Revenue"], label="Actual Revenue")
plt.plot(combined_df.index, combined_df["WindowAverage"], label="12 Month Avg", linestyle='dotted')
plt.plot(combined_df.index, combined_df["Naive"], label="Naive", linestyle='dotted')
plt.plot(combined_df.index, combined_df["SeasonalNaive"], label="Seasonal Naive", linestyle='dotted')

# chart formatting
plt.xlabel("Date")
plt.ylabel("Revenue")
plt.title("Benchmark Forecasting Results for US Ice Cream Revenue")
plt.legend()

# save the plot
# plt.savefig("chart1", dpi = 300, bbox_inches = "tight")
```

```python
# create a first order difference of the data
us_ic_diff1 = us_ic_clean.copy()
us_ic_diff1["Revenue_Diff"] = us_ic_diff1["Revenue"].diff(1)

# last value of the first order difference
last_value = us_ic_diff1["Revenue"].iloc[-1]

# set up models to train
models = [
    WindowAverage(window_size = 12),          # meanf
    Naive(),                                # naive
    SeasonalNaive(season_length=12),        # snaive
]

sf = StatsForecast(
    models=models,
    freq='ME'
)

# fit the model and forecast for 12 months ahead
Y_hat_diff1_df = sf.forecast(df = us_ic_diff1, 
                             time_col = "Date", 
                             target_col = "Revenue_Diff", 
                             id_col = "unique_id", 
                             h=12, 
                             fitted=False)

print(Y_hat_diff1_df.head())

# convert date to be first of the month
Y_hat_diff1_df["Date"] = Y_hat_diff1_df["Date"].dt.to_period("M").dt.to_timestamp()

# undifference the forecasted data for the first order difference
Y_hat_diff1_df["WindowAverage"] = Y_hat_diff1_df["WindowAverage"].cumsum() + last_value
Y_hat_diff1_df["Naive"] = Y_hat_diff1_df["Naive"].cumsum() + last_value
Y_hat_diff1_df["SeasonalNaive"] = Y_hat_diff1_df["SeasonalNaive"].cumsum() + last_value

# concat both df together
combined_diff1_df = pd.concat([us_ic_diff1, Y_hat_diff1_df], axis=0)

# make date the index
combined_diff1_df.set_index("Date", inplace=True)

print(combined_diff1_df.head())

# plot the combined data
plt.figure(figsize=(10, 6))

# plot the original revenue data as line and forecast as dotted line
plt.plot(combined_diff1_df.index, combined_diff1_df["Revenue"], label="Actual Revenue")
plt.plot(combined_diff1_df.index, combined_diff1_df["WindowAverage"], label="12 Month Avg", linestyle='dotted')
plt.plot(combined_diff1_df.index, combined_diff1_df["Naive"], label="Naive", linestyle='dotted')
plt.plot(combined_diff1_df.index, combined_diff1_df["SeasonalNaive"], label="Seasonal Naive", linestyle='dotted')

# chart formatting
plt.xlabel("Date")
plt.ylabel("Revenue")
plt.title("Benchmark Forecasting Results for US Ice Cream Revenue (First Order Difference)")
plt.legend()

# save the plot
# plt.savefig("chart2", dpi = 300, bbox_inches = "tight")
```
